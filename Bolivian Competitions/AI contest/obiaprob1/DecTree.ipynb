```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error, r2_score

# Supongamos que la columna objetivo se llama 'quality' y el resto son características
X = trn.drop('quality', axis=1)  # Características
y = trn['quality']  # Etiqueta objetivo

# Dividir el conjunto de datos en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
#test_size=2, rand


# Crear el modelo de Random Forest
model = DecisionTreeRegressor(random_state=1)
#n_estimators=100, rand


# Entrenar el modelo
model.fit(X_train, y_train)

# Hacer predicciones
y_pred = model.predict(X_test)

# Evaluar el modelo
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

# Imprimir resultados
print(f'Mean Squared Error: {mse:.2f}')
print(f'R^2 Score: {r2:.2f}')

# Asegúrate de que las columnas de prueba coincidan con las de entrenamiento
tst_features = tst # Eliminar la columna 'id'
model
# Hacer predicciones
predicciones_test = model.predict(tst_features)

# Crear un DataFrame para las predicciones
resultados = pd.DataFrame({'id': tst['id'], 'quality': predicciones_test})

# Guardar las predicciones en un archivo CSV
resultados.to_csv('predicciones.csv', index=False)
```
