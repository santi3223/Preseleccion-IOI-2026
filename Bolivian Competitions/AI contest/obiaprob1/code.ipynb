```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_squared_error
```

```python
X = trn.drop('quality', axis=1) 
y = trn['quality']
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
model = GradientBoostingRegressor(random_state=1)
model.fit(X_train, y_train)
```

```python

y_pred = model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)  # Calcular RMSE

print(f'Mean Squared Error: {mse:.2f}')
print(f'Root Mean Squared Error: {rmse:.2f}')  # Imprimir RMSE
```

```python
tst_features = tst
predicciones_test = model.predict(tst_features)

resultados = pd.DataFrame({'id': tst['id'], 'quality': predicciones_test})

resultados.to_csv('predicciones.csv', index=False)
```
